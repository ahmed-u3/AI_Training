```python
#Importations

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adamax, Nadam
```

### Dataset

```python
x= [[80.2, 300.5, 12.3, 1500, 2400, 5.5, 45.0, 0.85, 0.74, 0.65, 0],
    [95.1, 310.2, 15.8, 1600, 2750, 6.0, 40.0, 0.90, 0.80, 0.70, 1],
    [78.9, 290.7, 10.2, 1400, 2000, 5.2, 50.0, 0.80, 0.68, 0.60, 0],
    [102.3, 315.4, 20.0, 1700, 3000, 6.5, 55.0, 0.95, 0.82, 0.72, 2],
    [85.4, 305.1, 13.5, 1550, 2500, 5.7, 48.0, 0.88, 0.75, 0.68, 1]]

y= [0, 1, 0, 1, 0]
```

#### Convert X and Y into numpy array

```python
x = np.array(x)
x
```

```python
y = np.array(y)
y
```

## Model

#### Function to choose optimizer for the model

```python
def build_model(optimizer_choice):
    model = Sequential()
    model.add(Dense(32, input_dim=x.shape[1], activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model with the chosen optimizer
    model.compile(optimizer=optimizer_choice,
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# Map for optimizers
optimizers = {
    'SGD': SGD(),
    'Adam': Adam(),
    'RMSprop': RMSprop(),
    'Adagrad': Adagrad(),
    'Adadelta': Adadelta(),
    'Adamax': Adamax()
}
```

#### Training the models on each optimizer

```python
accuracy_map = {}
loss_map = {}
for name, optimizer in optimizers.items():
    print(f"\nTraining with {name} optimizer")
    model = build_model(optimizer)
    model.fit(x, y, epochs=20, batch_size=2)

    # Evaluate the model on the training data
    loss, accuracy = model.evaluate(x, y)
    accuracy_map[name] = accuracy
    loss_map[name] = loss
    print(f"Optimizer: {name}, Accuracy: {accuracy * 100}")
```

```python
accuracy_map
```

```python
loss_map
```

#### Comparing Accuracy and Loss

```python
plt.figure(figsize=(10, 6))
plt.bar(accuracy_map.keys(), accuracy_map.values())

# Adding title and labels
plt.title('Optimizer Accuracy Comparison', fontsize=16)
plt.xlabel('Optimizers', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.ylim(0, 1)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(loss_map.keys(), loss_map.values())

# Adding title and labels
plt.title('Optimizer Loss Comparison', fontsize=16)
plt.xlabel('Optimizers', fontsize=14)
plt.ylabel('Loss', fontsize=14)

plt.show()
```
